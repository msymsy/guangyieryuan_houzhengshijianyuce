```python
import pickle as pl
def load_data_all(filename):
    ##读取数据
    with open('源数据/pkl/'+filename+'.pkl', 'rb') as f:
        data_all = pl.load(f)
    return data_all  

def write_pkl(df,filename):
    #保存数据
    with open('源数据/pkl/'+filename+'.pkl', 'wb') as f:
        pl.dump(df,f,-1)
```

```python
#对年龄的处理
def age_rewrite(x):
    if 'Y' in x:
        sp=x.split('Y')
        if len(sp)==2:
            return int(sp[1])
    if 'M' in x:
        sp=x.split('M')
        if len(sp)==2:
            return int(sp[1])/12
    if ' ' in x:
        return 0
    return x
data2018['年龄']=data2018['年龄'].apply(lambda x:age_rewrite(x))
```
